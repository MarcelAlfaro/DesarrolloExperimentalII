```python
# Actividad 2 
# Elaborar un programa para colocar N partículas (puntos)
# en una recta de longitud L, de forma tal que la distancia 
# de separación entre ellas sea uniforme.

#Importamos librerias
import matplotlib.pyplot as plt

#Solicitamos cuantas particulas se graficaran 
n = input("Ingrese el número de particulas")

#Solicitamos la longitud L 
l = input ("Ingrese la longitud L")

#Creamos un paso para mantener los puntos equidistantes
n = int(n)
l = int(l)
h = l/n
rectay = []
rectax = []
#Creamos un array con el número de particulas 
for j in range(n):
    rectax.append (h*j)
    rectay.append (0)
```

```python
#Gráficamos el arreglo 
plt.figure(figsize = (20,5)) 
plt.plot(rectax ,rectay, "o")
```
